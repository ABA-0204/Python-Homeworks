```python
##ai
import threading
from collections import Counter
import os

def count_words(lines, counter_list, index):
    """Count word occurrences in a list of lines."""
    word_count = Counter()
    for line in lines:
        words = line.strip().lower().split()
        word_count.update(words)
    counter_list[index] = word_count

def read_file_in_chunks(filename, num_threads):
    """Read the file and divide lines into roughly equal chunks for threads."""
    with open(filename, 'r', encoding='utf-8') as f:
        lines = f.readlines()

    total_lines = len(lines)
    chunk_size = total_lines // num_threads
    chunks = []

    for i in range(num_threads):
        start = i * chunk_size
        end = None if i == num_threads - 1 else (i + 1) * chunk_size
        chunks.append(lines[start:end])

    return chunks

def count_words_in_file_threaded(filename, num_threads=4):
    """Main function to count words using multiple threads."""
    chunks = read_file_in_chunks(filename, num_threads)
    threads = []
    counter_list = [None] * num_threads

    for i in range(num_threads):
        t = threading.Thread(target=count_words, args=(chunks[i], counter_list, i))
        threads.append(t)
        t.start()

    for t in threads:
        t.join()

    total_count = Counter()
    for counter in counter_list:
        total_count.update(counter)

    return total_count

# Example usage
if __name__ == "__main__":
    filename = "large_text_file.txt"  # Replace with your actual file path
    num_threads = 4

    if os.path.exists(filename):
        word_counts = count_words_in_file_threaded(filename, num_threads)
        print("Top 20 most common words:")
        for word, count in word_counts.most_common(20):
            print(f"{word}: {count}")
    else:
        print(f"File '{filename}' not found.")
```
