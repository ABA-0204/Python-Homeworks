```python
import numpy as np
```

```python
# From a Python list
arr1 = np.array([1, 2, 3])  # 1D array
print(arr1)
# 2D array (matrix)
arr2 = np.array([[1, 2, 3], [4, 5, 6]])
print (arr2)
# Array of zeros
zeros = np.zeros((2, 3))
print(zeros)
# Array of ones
ones = np.ones((3, 2))
print(ones)
# Array with a range of values
range_arr = np.arange(0, 10, 2)
print(range_arr)
# Random values
random_arr = np.random.rand(2, 3)
print(random_arr)
```

```python
arr = np.array([1, 2, 3])
print(arr.dtype)  # int64 or int32 depending on system

# You can set the data type
arr_float = np.array([1, 2, 3], dtype='float32')
print(arr_float.dtype)  # float32
```

```python
# Write a NumPy program to create a null vector of size 10 and update the sixth value to 11.
zeros=np.zeros(10)
print(zeros)
zeros[5]=11
print(zeros)
```

```python
# Write a NumPy program to create an array with values ranging from 12 to 38.
range_arr=np.arange(12,38)
print(range_arr)
```

```python
#Write a NumPy program to convert an array to a floating type.
float_array=np.array([4,5,6,7],dtype='float32')
print(float_array)
print(type(float_array))
```

```python
#Write a NumPy program to convert Centigrade degrees into Fahrenheit degrees. Centigrade values are stored in a NumPy array.
centigrade = np.array([0, 20, 37, 100])

# Convert to Fahrenheit
fahrenheit = centigrade * 9/5 + 32

# Print the result
print("Centigrade:", centigrade)
print("Fahrenheit:", fahrenheit)
```

```python
# Write a NumPy program to append values to the end of an array.
original_arr=np.array([10,20,30])
append_value=[40,50,60,70,80,90]
new_arr=np.append(original_arr, append_value)
print(new_arr)
```

```python
#Create a random NumPy array of 10 elements and calculate the mean, median, and standard deviation of the array.

random_array = np.random.rand(10)  

mean = np.mean(random_array)
median = np.median(random_array)
std_dev = np.std(random_array)

print("Random array:", random_array)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
```

```python
#Create a 10x10 array with random values and find the minimum and maximum values.
random_array = np.random.rand(10,10)  
Min_value=np.min(random_array)
MAx_value=np.max(random_array)
print('the minimum value in random array is :', Min_value)
print('The max value in random_array is :',MAx_value)
```
